```python
import pandas as pd
import glob
```

```python
lista_arquivos = glob. glob ('../base_dados/*.csv') #bliblioteca glob com método glob, o consulta*-->vai abrir todos os arquivos que começam com consulta
print(lista_arquivos)
```

```python
lista_df = []
```

```python
#loop para a lista
for csv in lista_arquivos:
    df= pd.read_csv(csv, sep=';', encoding='latin_1')
    lista_df.append(df)
```

```python
print(lista_df)
```

```python
candidaturas_PE_RN = pd.concat(lista_df, axis=0, ignore_index=True)
```
